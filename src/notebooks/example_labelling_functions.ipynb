```python
def process_nounPhrases(self, docTxt,
                            stopwordSet,
                            customizedStopwordSet,
                            generalNPSet,
                            non_stemming_set,
                            lemma_normalize):
        '''
        extract noun phrases from each document
        :param docTxt:
        :param stopwordSet:
        :param customizedStopwordSet:
        :param generalNPSet:
        :param non_stemming_set:
        :param nlp_spacy:
        :return:
        '''
        phrases = set()
        chunkDoc = self.nlp_spacy_model(docTxt)
        for chunk in chunkDoc.noun_chunks:
            currentChunk = chunk.text.lower().strip().replace("\n", " ")

            if currentChunk.strip() != "":
                currentChunk = currentChunk.translate({ord(c): " " for c in "!@#$%^&*ω()»¿'[]{};:,./<>?\|`~°=\"+"})

            phrase = self.remove_terms(stopwordSet, currentChunk)
            phrase = self.remove_digit_unit(phrase)

            if lemma_normalize:
                phrase = self.apply_lemmatization(phrase)

            if len(phrase.strip().split()) == 1:
                phrase = self.remove_terms(customizedStopwordSet, phrase)

            if phrase.strip() != "":
                phrases.add(phrase.strip())

        return phrases
```

```python
def get_np_present_in_full_text(df_full_text):
    c_in_ft=0
    for i, e in enumerate(data.dict_der_text):
        
        np_der=set(get_np(data.dict_der_text[e], True))
        df_temp=df_full_text.loc[df_full_text['PPID'] == e]
        
        abst=df_temp['ABEN'].iloc[0]
        claim=df_temp['CLMEN'].iloc[0]   
        detden=df_temp['DETDEN'].iloc[0] 
    
        np_abst=get_np(abst, True)
        np_claim=get_np(claim, True)
        np_detden=get_np(detden,True)
        
        set_all_np_full_text=set(itertools.chain(np_abst,np_claim,np_detden))
        c_in_ft+=len(np_der.intersection(set_all_np_full_text))
        
        if i%10==0:
            print(i, np_der.intersection(set_all_np_full_text))
            
    print(c_in_ft, c_in_ft/100)
```

```python
def get_np_do_not_present_in_full_text(df_full_text):
    c_not_in_ft=0
    for i, e in enumerate(data.dict_der_text):
        np_der=set(get_np(data.dict_der_text[e], False))
        
        df_temp=df_full_text.loc[df_full_text['PPID'] == e]
        
        abst=df_temp['ABEN'].iloc[0]
        claim=df_temp['CLMEN'].iloc[0]   
        detden=df_temp['DETDEN'].iloc[0] 
    
    
        np_abst=get_np(abst, False)
        np_claim=get_np(claim, False)
        np_detden=get_np(detden,False)
        
        set_all_np_full_text=set(itertools.chain(np_abst,np_claim,np_detden))
        c_not_in_ft+=len(np_der-set_all_np_full_text)
        
        if i%10==0:
            print(i, np_der-set_all_np_full_text)
            
    print(c_not_in_ft, c_not_in_ft/100)
```
